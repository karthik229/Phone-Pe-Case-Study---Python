# Task 3: Data Quality Checks

# 3.1: Ensure data consistency across state and district levels

1. For each state, calculate the total number of transactions, total transaction amount, and total registered users by summing up the values from the district level data.

```python
state_level_summary = district_txn_users_df.groupby('State').agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum'
}).reset_index()

display(state_level_summary)
```

2. Compare the results with the corresponding values at the state level to ensure they match.

```python
state_summary = state_txn_users_df.groupby('State').agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum',
    'Registered Users': 'sum'
}).reset_index()

comparison_df = pd.merge(
    state_level_summary, state_summary,
    on = 'State',
    suffixes=('_District', '_State')
)

comparison_df['Transactions_Discrepancy'] = comparison_df['Transactions_District'] - comparison_df['Transactions_State']
comparison_df['Amount_Discrepancy'] = comparison_df['Amount (INR)_District'] - comparison_df['Amount (INR)_State']
comparison_df['Users_Discrepancy'] = comparison_df['Registered Users_District'] - comparison_df['Registered Users_State']

display(comparison_df[['State', 'Transactions_Discrepancy', 'Amount_Discrepancy', 'Users_Discrepancy']])
```

3. Display any discrepancies found between the district-level and state-level data.

```python
discrepancies = comparison_df[
    (comparison_df['Transactions_Discrepancy'] !=0) |
    (comparison_df['Amount_Discrepancy'] !=0) |
    (comparison_df['Users_Discrepancy'] !=0)
]

print("Discrepancies found:")
display(discrepancies[['State', 'Transactions_Discrepancy', 'Amount_Discrepancy', 'Users_Discrepancy']])
```
